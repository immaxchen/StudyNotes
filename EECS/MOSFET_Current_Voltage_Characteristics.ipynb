```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
C = 0.21
L = 0.01
VT = 1.0
```

```python
def ID_lin(VG, VD):
    return C * ( 2*(VG-VT)*VD - VD**2 )
```

```python
def ID_sat(VG, VD):
    return C * (VG-VT)**2 * ( 1 + L*(VD-(VG-VT)) )
```

```python
def ID_edge(VD):
    # VD == VG-VT
    return C * (VD)**2
```

```python
VD = np.arange(0,6,0.1)

for VG in range(3,6):
    plt.plot(VD, ID_lin(VG, VD))
    plt.plot(VD, ID_sat(VG, VD))
    
plt.plot(VD, ID_edge(VD), ":")

kwargs = dict(xytext=(20,10), textcoords="offset points", arrowprops=dict(arrowstyle="->"))

plt.annotate(s="VD=VG-VT=2", xy=(2, ID_edge(2)), **kwargs)
plt.annotate(s="VD=VG-VT=3", xy=(3, ID_edge(3)), **kwargs)
plt.annotate(s="VD=VG-VT=4", xy=(4, ID_edge(4)), **kwargs)
plt.annotate(s="VD=2, VG=5", xy=(2, ID_lin(5,2)), **kwargs)
plt.annotate(s="VD=4, VG=3", xy=(4, ID_sat(3,4)), **kwargs)

plt.gca().set(xlim=(0,6), ylim=(0,4))
plt.xlabel("VD")
plt.ylabel("ID")
plt.show()
```
